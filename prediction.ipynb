```python
%reset -f

import pickle
import cv2
import os
```

```python
from keras.models import load_model

model = load_model('saved_models/pnemonia_prediction.h5')
```

```python
CLASS = ['NORMAL','PNEUMONIA']

def predict_img(url):
    
    arr = cv2.imread(url)
    arr = cv2.resize(arr, (100, 100))
    arr = arr.reshape(-1, 100, 100, 3)
    arr = arr/255
    
    prediction = model.predict([arr])
    # print(prediction)
    # print(CLASS[prediction.argmax()])
    
    return prediction
```

```python
# NORMAL
n = 0
m = 0
for filename in os.listdir(r'data/test/NORMAL/'):
    m += 1
    prediction = predict_img(r'data/test/NORMAL/'+str(filename))
    if CLASS[prediction.argmax()] == "NORMAL":
        n += 1
print("n = ",n)
print("m = ",m)
# predict_img(r'data/test/NORMAL/IM-0010-0001.jpeg')
```

```python
# PNEUMONIA
p = 0
q = 0
for filename in os.listdir(r'data/test/PNEUMONIA/'):
    q += 1
    prediction = predict_img(r'data/test/PNEUMONIA/'+str(filename))
    if CLASS[prediction.argmax()] == "PNEUMONIA":
        p += 1
print("p = ",p)
print("q = ",q)

# predict_img(r'data\test\PNEUMONIA\person17_virus_48.jpeg')
```
