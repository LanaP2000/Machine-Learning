```python
%reset -f

import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import os
```

1 - PCA on test/NORMAL dataset

```python
# loading images
# os.chdir("../../../")
loaded_images_NORMAL = list()
for filename in os.listdir(r'data/test/NORMAL/'):
    # load image
    img_data = cv2.imread(r'data/test/NORMAL/' + filename)
    # store loaded image
    loaded_images_NORMAL.append(img_data)
    # print('> loaded %s %s' % (filename, img_data.shape))
```

```python
# Application of PCA
compressed_images_NORMAL = list()

# Splitting image in arrays 
for img in loaded_images_NORMAL:
    B,G,R = cv2.split(img)
    # Applying PCA for the 20 first principal components
    pca = PCA(20)

    # Red channel
    R_transformed = pca.fit_transform(R)
    R_inverted = pca.inverse_transform(R_transformed)

    # Green channel
    G_transformed = pca.fit_transform(G)
    G_inverted = pca.inverse_transform(G_transformed)

    # Blue channel
    B_transformed = pca.fit_transform(B)
    B_inverted = pca.inverse_transform(B_transformed)

    # Compress image
    compressed_images_NORMAL.append((np.dstack((R_inverted,G_inverted,B_inverted))).astype(np.uint8))
    
```

```python
# Saving images in directory
i = 0
os.chdir(r'data/test/NORMAL_PCAcompressed/')
for comp_img in compressed_images_NORMAL:
    cv2.imwrite('IM'+str(i)+'.jpeg',comp_img)
    i += 1
```

```python
# Display an image for testing
img = compressed_images_NORMAL[105]
plt.imshow(img)
```
